**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
pd.set_option('display.max_columns',None)
```

```python
# Read the CSV file
file_path = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the data
df.head()
```

```python
import numpy as np
df_we_mean = df['WorkExp'].mean()
df['WorkExp'].replace(np.nan, df_we_mean,inplace = True) 
```

```python
df['WorkExp']
```

```python
df['Age'].unique()
```

```python
age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 30,
    'Prefer not to say': 35,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}
```

```python
df['Age'] = df['Age'].map(age_map)
```

```python
df['Age'].dtypes
```

```python
df['TimeAnswering'].value_counts()
```

```python
time_map = {
    '15-30 minutes a day': 22.5,
    '30-60 minutes a day ': 45,
    'Less than 15 minutes a day': 10,
    '60-120 minutes a day': 90,
    'Over 120 minutes a day': 125
}
```

```python
df['TimeAnswering'] = df['TimeAnswering'].map(time_map)
```

```python
df['TimeAnswering'].replace(np.nan,45,inplace=True)
```

```python
df['TimeSearching'] = df['TimeSearching'].map(time_map)
```

```python
df['TimeSearching'].replace(np.nan,45,inplace=True)
```

```python
df['TimeSearching']
```

```python
df['CompTotal'].mean()
```

```python
df_ct_mean = df['CompTotal'].mean()
df['CompTotal'].replace(np.nan,df_ct_mean,inplace=True)
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
QUERY = """
SELECT CompTotal 
FROM main

"""
df = pd.read_sql_query(QUERY,conn)

sns.displot(df,bins = 5)
```

**Box Plots**

Plot a box plot of Age.


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
df = pd.read_sql_query(QUERY, conn)
plt.rcParams['figure.figsize'] = [15,8]

ax = sns.boxplot(data = df, orient = 'h')
ax.set_title('Age Distribution', size = 18)

plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
df['WorkExp']
```

```python
## Write your code here
QUERY = """
SELECT Age, COUNT(*) as count, WorkExp
FROM main
GROUP BY Age
ORDER BY Age
"""
df = pd.read_sql_query(QUERY, conn)
plt.rcParams['figure.figsize'] = [15,8]

sns.scatterplot(data = df, x = 'Age', y = 'WorkExp')

plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here
QUERY = """
SELECT Age, TimeSearching, Frustration
FROM main

"""
df = pd.read_sql_query(QUERY,conn)
df
plt.rcParams['figure.figsize'] = [15,15]

sns.scatterplot(data = df, y = 'TimeSearching', x = 'Frustration', size = 'Age', alpha = 0.6, sizes=(1, 1000))
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
QUERY = """
SELECT *, COUNT(*) AS count 
FROM main
GROUP BY DatabaseWantToWorkWith
ORDER BY count DESC
LIMIT 5

"""
df = pd.read_sql_query(QUERY,conn)
df
data = df['count']
labels = df['DatabaseWantToWorkWith']

plt.rcParams['figure.figsize'] = [8,8]

fig, ax = plt.subplots()
ax.pie(data, labels = labels, autopct='%1.1f%%', explode = (0.1, 0, 0, 0, 0), shadow = True)
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
QUERY = """
SELECT TimeSearching, TimeAnswering, Age
FROM main
WHERE Age BETWEEN 30 AND 35

"""
df = pd.read_sql_query(QUERY,conn)
df.value_counts()


time_searching = df['TimeSearching'].quantile(0.5)
time_answering = df['TimeAnswering'].quantile(0.5)

plt.rcParams['figure.figsize'] = [8,8]

fig, ax = plt.subplots()

ax.bar('Age 30 - 35', time_searching, label='TimeSearching')
ax.bar('Age 30 - 35', time_answering , bottom=time_searching, label='TimeAnswering')
ax.legend()

plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
df['CompTotal'].mean()
```

```python
## Write your code here
QUERY = """
SELECT CompTotal, Age
FROM main
WHERE Age BETWEEN 45 AND 60

"""
df = pd.read_sql_query(QUERY,conn)

medians = df.groupby('Age')['CompTotal'].median()

data = pd.DataFrame(list(zip(medians.index, medians.values)),
               columns =['Age', 'MedCompTotal'])

plt.rcParams['figure.figsize'] = [8,8]

sns.lineplot(data=data, x='Age', y = 'MedCompTotal')

plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
QUERY = """
SELECT MainBranch, COUNT(*) AS Frequency
FROM main
GROUP BY MainBranch

"""
df = pd.read_sql_query(QUERY,conn)

plt.rcParams['figure.figsize'] = [14, 8]

sns.barplot(data = df, y = 'MainBranch', x = 'Frequency', orient = 'h')

plt.show()
```
